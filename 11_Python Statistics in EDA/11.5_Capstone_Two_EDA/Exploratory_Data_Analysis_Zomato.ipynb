```python
df_country = pd.read_excel('data/Country-Code.xlsx')
df_country.head()
```

```python
# Merge datasets
df=pd.merge(df1, df_country, on= 'Country Code', how= 'left')   #so we can get a final df with country name
df.head(2)
```

```python
df.isnull().sum()
```

```python
df.drop(['Restaurant ID', 'Country Code', 'Locality Verbose', 'Longitude', 'Latitude'],axis=1,
inplace=True)
```

```python
df.head(2)
```

## Step 3: Inferential Statistics
1. Perform inferential statistics to gain insights into your data
2. Identify any abnormal or unusual results in your dataset
3. Formulate null and alternative hypothesis to test based on your understanding of the data and main question.

```python
# Top 3 countries that uses zomato
top_3 = df['Country'].value_counts()[:3]
top_3
```

```python
# Plot pie chart
plt.pie(top_3,labels=top_3.index)
circle = plt.Circle( (0,0), 0.5, color='white')    #Creating a Circle
p=plt.gcf()                                        #Adding circle to the Pie Chart
p.gca().add_artist(circle)
plt.title("Top 3 countries that uses zomato")      #Add title

plt.show()
```

```python
# Top 10 Countries that uses zomator
top_10 = df['Country'].value_counts()[:10].rename_axis('Country').reset_index(name='count')
top_10.style.background_gradient(cmap='Blues')
```

```python
# Countries with minimum interaction
df['Country'].value_counts()[-3:].rename_axis('Country').reset_index(name='count').style.background_gradient(cmap='Blues')
```

```python
# Most popular cuisines
popular_cuisines = df['Cuisines'].value_counts()[:10].rename_axis('Cuisines').reset_index(name='count')
popular_cuisines.style.background_gradient(cmap='Blues')
```

## Step 4: Explore Data Relationships
1. Use data visualization techniques to explore relationships between variables and compare them to the response variable.
2. Utilize matplotlib and seaborn functions to create various plots, such as scatter plots, line plots, histograms, or categorical plots, to visualize distributions, correlations, or group comparisons.

```python
# Plot the most popular cuisines
plt.title('Most popular cuisines')

sns.scatterplot(y=popular_cuisines['Cuisines'], x=popular_cuisines['count'], hue=popular_cuisines['Cuisines'])
```

The most popular are Chinese and North Indian cuisines.

```python
# Analyzing "Average Cost for two" column
by_rating = df.groupby(['Rating color']).mean().round()
by_rating
```

```python
plt.title('Aggregate rating vs Average Cost for two')
sns.barplot(x=by_rating['Aggregate rating'], y=by_rating['Average Cost for two'])
```

Top rated restaurants have high average cost

```python
# Analyzing Rating text column

plt.title('Rating text vs Average Cost for two')
sns.barplot(x = df['Rating text'], y=df['Average Cost for two'], palette = ['cyan', 'orange', 'red', 'green', 'blue', 'grey'])
```
