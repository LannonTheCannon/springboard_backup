```python
df_country = pd.read_excel('data/Country-Code.xlsx')
df_country.head()
```

```python
# Merge datasets
df=pd.merge(df1, df_country, on= 'Country Code', how= 'left')   #so we can get a final df with country name
df.head(2)
```

```python
df.drop(['Restaurant ID', 'Country Code', 'Locality Verbose', 'Longitude', 'Latitude'],axis=1,
inplace=True)
```

## Step 3: Inferential Statistics
1. Perform inferential statistics to gain insights into your data

```python
# Top 3 countries that uses zomato
top_3 = df['Country'].value_counts()[:3]
top_3
```

```python
# Plot pie chart
plt.pie(top_3,labels=top_3.index)
circle = plt.Circle( (0,0), 0.5, color='white')    #Creating a Circle
p=plt.gcf()                                        #Adding circle to the Pie Chart
p.gca().add_artist(circle)
plt.title("Top 3 countries that uses zomato")      #Add title

plt.show()
```

```python
# Top 10 Countries that uses zomator
top_10 = df['Country'].value_counts()[:10].rename_axis('Country').reset_index(name='count')
top_10.style.background_gradient(cmap='Blues')
```

```python
# Countries with minimum interaction
df['Country'].value_counts()[-3:].rename_axis('Country').reset_index(name='count').style.background_gradient(cmap='Blues')
```

```python
# Most popular cuisines
popular_cuisines = df['Cuisines'].value_counts()[:10].rename_axis('Cuisines').reset_index(name='count')
popular_cuisines.style.background_gradient(cmap='Blues')
```
